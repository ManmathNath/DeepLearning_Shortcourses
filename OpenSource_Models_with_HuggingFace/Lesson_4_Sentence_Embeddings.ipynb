```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python

sentence1 = ['the cat sits outside','who are you ?','I am a boy']
sentence2 = ['A women watches Tv','how are you ?','I am superman']
embedding1 = model.encode(sentence1 ,convert_to_tensor = True)
embedding2 = model.encode(sentence2 ,convert_to_tensor = True)
```

```python
print(embedding1)
```

```python
from sentence_transformers import util
cosine_scores = util.cos_sim(embedding1, embedding2)
```

```python
print("cosine_scores:",cosine_scores)
```

```python
for i in range(len(sentence1)):
    print("{}\t\t{}\t\t Scores:{:.4f}".format(sentence1[i],sentence2[i],cosine_scores[i][i]))
```
