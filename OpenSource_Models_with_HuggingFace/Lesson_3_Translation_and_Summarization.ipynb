```python
from transformers import pipeline
import torch
```

```python
translator = pipeline(task = "translation",
                     model = "facebook/nllb-200-distilled-600M",torch_dtype = torch.bfloat16)
```

```python
text ="""my puppy is adorable.\
         your kitten is cute.\
         her panda is friendly.\
      """
```

```python
text_translated= translator(text,src_lang = "eng_Latn",tgt_lang = "hin_Deva")
```

```python
text_translated
```

#TEXT_SUMMARIZATION

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn",torch_dtype = torch.bfloat16)
```

```python
text ="""In the heart of the woods, a traveler, uncertain and weary, sought guidance from a venerable owl perched atop a gnarled oak.

          With keen eyes, the owl imparted wisdom: "Trust your inner compass; it knows the path you seek."

         "Embrace the winds of change; they carry the seeds of growth," the owl continued, its voice echoing through the forest.

        With newfound clarity, the traveler bid farewell, carrying the owl's timeless advice in heart and mind."""
```

```python
summary = summarizer(text,min_length = 10,max_length = 100)
```

```python
summary
```
